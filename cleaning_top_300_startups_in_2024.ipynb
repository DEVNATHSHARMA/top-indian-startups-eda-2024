```python
def convert_to_crores(value):
    value = str(value).strip()
    if value.startswith('₹'):
        # Remove ₹ and convert to float
        rupees=float(value.replace('₹', '').replace(',', '').strip())
        return rupees/10000000
    elif value.startswith('$'):
        # Remove $ and convert to INR
        usd_value = float(value.replace('$', '').replace(',', '').strip())

        return (usd_value * 85)/10000000

```

```python
df['Total_funding(in crores)'] = df['Total Funding'].apply(convert_to_crores)
```
